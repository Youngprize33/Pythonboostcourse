# 데이터 전처리

## 기간에서 연도를 분리하기

```python
# 연도 추출
int("2019 4/4 p)".split()[0])
```

```python
df["연도"] = df["기간"].map(lambda x : int(x.split()[0]))
df.head()
```

```python
# 기간에서 분기만 분리하기
"2019 4/4 p)".split()[1].split("/")[0]
```

```python
df["분기"] = df["기간"].map(lambda x : int(x.split()[1].split("/")[0]))
df.head()
```

## 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기

```python
# - 문자를 결측치로 변경하고 float 타입으로 변경
df["백만원"] = df["백만원"].replace("-", pd.np.nan).astype(float)
df.head()
```

## 필요없는 데이터 제거하기

```python
df.info()
```

```python
# 합계 데이터는 따로 구할 수 있기 때문에 전체 데이터에서 제거
# 연산자 조건에 대해서 묶였을 때 잘 계산할것
# 용량 비교해보기
df = df[(df["국가(대륙)별"] != "합계" ) & (df["상품군별"] != "합계")].copy()
```

```python
# info로 정보보기
df.info()
```

```python
# 결측치 보기
df.isnull().sum()
```

# K-Beauty 시각화

## 전체 상품군 판매액

```python
# 판매유형별 데이터는 일부 기간에 "계"만 존재
# 판매유형별 == "계" 데이터만 가져와서 봐야
# 평균 값을 구하는 그래프에서 올바른 값을 표현 가능
df_total = df[df["판매유형별"] == "계"].copy()
df_total.head()
```

```python
# 연도, 판매액 lineplot
sns.lineplot(data = df_total, x="연도", y="백만원")
```

* Legend를 그래프 밖에 그리는 방법 - https://stackoverflow.com/questions/30490740/move-legend-outside-figure-in-seaborn-tsplot

```python
# 연도, 판매액 lineplot으로 그리고 상품군별 다른 색상 표기
sns.lineplot(data = df_total, x="연도", y="백만원", hue="상품군별")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# 위 그래프를 자세히 보기 위해 서브플롯 표시
sns.relplot(data=df_total, x="연도", y="백만원",
            hue="상품군별", kind="line", col="상품군별", col_wrap=4)
```

```python
# isin을 사용해 화장품만 제외를 하고 df_sub 변수에 담기
df_sub = df_total[-df_total["상품군별"].isin(["화장품", "의류 및 패션 관련상품"])].copy()
```

```python
# 연도별 판매액을 상품군별로 relplot을 활용해 서브플롯 그리기
sns.relplot(data=df_sub, x="연도", y="백만원",
            hue="상품군별", col="상품군별", col_wrap=4, kind="line")
```

## 화장품의 온라인쇼핑 해외직접판매

```python
# df_cosmetic 변수에 상품군별 화장품 데이터만 가져오기
df_cosmetic = df_total[df_total["상품군별"] == "화장품"].copy()
df_cosmetic["상품군별"].unique()
```

```python
# 연도와 판매액을 lineplot으로 그리고 분기별로 다른 색상 표현
plt.figure(figsize=(15,4))
sns.lineplot(data=df_cosmetic, x="연도", y="백만원", hue="분기")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x="기간", y="백만원")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화, "국가(대륙)별"로 다른 색상 표기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x="기간", y="백만원", hue="국가(대륙)별")
```

```python
# 중국 빼고 그려보기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic[df_cosmetic["국가(대륙)별"] != "중국"],
             x="기간", y="백만원", hue="국가(대륙)별")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화, "판매유형별"로 다른 색상 표기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[df["판매유형별"] != "계"].copy()
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별")
```

```python
# 면세점 빼고 보기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[(df["판매유형별"] != "계") & (df["판매유형별"] != "면세점")].copy()
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별", ci=None)
```

## 의류 및 패션관련 상품 온라인쇼핑 해외직접판매액

```python
# df_fashion 변수에 의류 데이터만 따로 담기
df_fashion = df[(df["상품군별"] == "의류 및 패션 관련상품") & (df["판매유형별"] == "계")].copy()
df_fashion.head()
```

```python
# "의류 및 패션관련 상품" 판매액에 대한 기간별 금액 데이터 시각화
# "국가(대륙)별"로 다른 색상 표기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x="기간", y="백만원", hue="국가(대륙)별")
```

```python
# "의류 및패션관련 상품" 판매액에 대한 기간별 금액 데이터 시각화
# "판매유형별"로 다른 색상 표기
df_fashion2 = df[(df["상품군별"] == "의류 및 패션 관련상품") & (df["판매유형별"] != "계")].copy()

plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2, x="기간", y="백만원", hue="판매유형별", ci=None)
```
